# Data Analysis


```python
#creating a new dataset of recent data from the main dataset
recent_films=film_titles.loc[film_titles['start_year'] >=2023]
recent_films.fillna(0)
```

```python
#Identifying the types of films present
for genres in recent_films['genres'].unique():
    print (genres)
```

In the Above, we have acknowledged the types of films that are recently getting released.

Accessing films with high ratings

```python
High_ratings=film_ratings.loc[film_ratings['numvotes']>=1000000]
High_ratings
```

```python
#getting the film with the highest number of votes
film_ratings.loc[film_ratings['numvotes']>=1000000].max()
```

```python
#getting the film with the lowest number of votes
film_ratings.loc[film_ratings['numvotes']>=1000000].min()
```

Acquiring film names with high ratings using the 'tconst' column

```python
Highrated_films=recent_films.loc[recent_films['tconst']>='tt0816692']
Highrated_films
```

Acquiring film names with higher domestic gross revenue

```python
high_gross_film=bom_movie_gross.loc[bom_movie_gross['domestic_gross']>600000000]
high_gross_film
```

Gross according to recent years

```python
bom_movie_gross.loc[bom_movie_gross['year']>=2018]
```

# Data Visualization

```python
import matplotlib.pyplot as plt
%matpotlib inline
```

Visualizing the domestic gross revenue from the year 2012 to 2018

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
y=high_gross_film['domestic_gross']
x=high_gross_film['year']
plt.xlabel(' Year')
plt.ylabel('Domestic Gross')
plt.title('Domestic Gross')

plt.plot(x,y)
plt.show()
```

Visualizing data of movies with higher ratings

```python
x=High_ratings['numvotes']
y= High_ratings['averagerating']
plt.scatter(x,y)
plt.xlabel('Number of Votes')
plt.ylabel('Average Rating')
plt.title('High Film Ratings')
plt.show()

```
